```python
import os
import shutil

def move_bmp_images(source_folder, target_folder):
    # Create the target folder if it doesn't exist
    os.makedirs(target_folder, exist_ok=True)

    # Iterate over all files in the source folder
    for filename in os.listdir(source_folder):
        if filename.endswith(".bmp"):
            # Construct the full file path
            source_path = os.path.join(source_folder, filename)
            target_path = os.path.join(target_folder, filename)
            
            # Move the file
            shutil.move(source_path, target_path)
            print(f'Moved: {filename}')

# Define the source and target folders
source_folder = 'C:\\Users\Dell\\Desktop\\machine learnining\\dataset'
target_folder = 'C:\\Users\\Dell\\Desktop\\machine learnining\dataset\\Imsges'

# Call the function to move .bmp images
move_bmp_images(source_folder, target_folder)
```
